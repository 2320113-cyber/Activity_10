```python
from openai import OpenAI
from langchain_community.chat_message_histories import ChatMessageHistory

# Setup the Local Client
client = OpenAI(
    base_url="http://127.0.0.1:1234/v1",  # include /v1
    api_key="lm-studio"
)

history = ChatMessageHistory()

print("Chatbot initialized for CJ Laguerta.")
print("Type 'exit' anytime to stop.\n")

while True:
    query = input("\n👉 Enter your message here: ")

    if query.lower() == "exit":
        print("\nExiting...")
        print("Program Finished.")
        break

    if not query.strip():
        continue

    print(f"\nUser Input: {query}")
    print("- " * 50)

    history.add_user_message(query)

    messages_with_history = [
        {"role": "user" if m.type == "human" else "assistant", "content": m.content}
        for m in history.messages
    ]

    try:
        stream = client.chat.completions.create(
            model="google/gemma-4-e4b",  # use the correct model name
            messages=messages_with_history,
            stream=True,
            max_tokens=256
        )

        full_ai_response = ""
        print("\nAI Response:\n")

        for chunk in stream:
            if chunk.choices[0].delta.content:
                content = chunk.choices[0].delta.content
                print(content, end="", flush=True)
                full_ai_response += content

        history.add_ai_message(full_ai_response)
        print("\n" + "=" * 100 + "\n")

    except Exception as e:
        print(f"\n❌ Error: {e}")
        print("=" * 100 + "\n")
```
